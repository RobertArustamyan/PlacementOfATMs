```python
import folium
from utils.atm_data_reading import get_atm_data
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# Define the map centered around Leoben
leoben_coordinates = (47.38, 15.095)

atm_map = folium.Map(location=leoben_coordinates,zoom_start=15)
```

```python
# Load ATM data from CSV
test_data_name = 'Leoben-73ATMsTEST.csv'
atms_data = get_atm_data(fr"{os.getenv('ATMS_DATA_PATH')}\{test_data_name}")

for atm in atms_data:
    folium.Marker(location=(atm.latitude,atm.longitude),popup=atm.name).add_to(atm_map)

# Display the generated map
atm_map
```
