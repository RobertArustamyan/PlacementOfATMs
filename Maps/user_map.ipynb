```python
import folium
from utils.generate_user_coordinates import generate_users_in_quadrilateral
from utils.users_living_quadrilateral_reading import get_living_quadrilateral_data
from utils.ordering import order_point_in_dictionary
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# Define the map centered around Leoben
leoben_coordinates = (47.38, 15.095)
# Creating map starting in leoben
user_map = folium.Map(location=leoben_coordinates, zoom_start=15)
```

```python
# Loading dataset of bounds
test_data_name = 'LeobenUserLivingSquares3Test.csv'
data = get_living_quadrilateral_data(fr"{os.getenv('USERS_DATA_PATH')}\{test_data_name}")

data = order_point_in_dictionary(data)
```

```python
for i, bounds in data.items():
    # Comment the line below to not generate users and see only bounds!
    generated_users = generate_users_in_quadrilateral(bounds, 50)
    for bound in bounds:
        folium.Marker(location=bound, icon=folium.Icon(color="red"),popup=i).add_to(user_map)

    # Generated user adding on map. Comment following 2 lines to not see on the map
    # for user in generated_users:
    #     folium.Marker(location=(user.latitude,user.longitude)).add_to(user_map)

    folium.PolyLine(locations=bounds + [bounds[0]], color="blue", weight=2.5, opacity=1).add_to(user_map)

```
